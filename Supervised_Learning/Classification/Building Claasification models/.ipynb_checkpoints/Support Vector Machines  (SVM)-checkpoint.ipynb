# Support Vector Machines (SVM)

SVM is a Machine Learning Algorithm which can be used for both classification and regression. The main concept behind the SVM is to plot each feature as aposition in an N dimensional space with the result of each feature being the the value of the coordinate it is in.  

When the different instances of the features have been put into their different segments then a line is drawn across the features thus seperating them and when new foreign data isn provided then the model then places the new data instance on a segment according to it's coordinate.

## Kernel

This is simply a function employed by the SVM to transform any seemingly inseperable problem into a seperable one.

It can be in four forms:

#### Linear

#### Polynomial

#### RBF

#### Sigmoid

## Dataset

We wil make use of scikit-learn's inbuilt Iris dataset.The dataset has 4 features and we will use the SVM to seperate the different types of IRIS plants. 

### Preparing and viewing the Data 

```python
#importing essential packages

import pandas as pd

import numpy as np

from sklearn.svm import SVC

from sklearn.datasets import load_iris

%matplotlib inline

import matplotlib.pyplot as plt
```

```python
#loading data 
iris = load_iris()

#features
x = iris.data[:, :2]

y = iris.target
```

Now we will view the data using matplotlib

```python
#we are going top create a mesh to plot
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1

y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1

h = (x_max / x_min) / 100

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

x_plot = np.c_[xx.ravel(), yy.ravel()]

#now for the regularization parameter
c = 1.0
```

#### SVC with Linear Kernel

```python
linear_kernel_model = SVC(kernel = 'linear', C=c, decision_function_shape='ovr')

linear_kernel_model.fit(x, y)

z = linear_kernel_model.predict(x_plot)

z = z.reshape(xx.shape)

plt.figure(figsize = (15, 5))

plt.subplot(121)

plt.contourf(xx, yy, z, cmap = plt.cm.tab10, alpha = 0.3)

plt.scatter(x[:, 0], x[:, 1], c = y, cmap = plt.cm.Set1)

plt.xlabel('Sepal Length')

plt.ylabel('Sepal Width')

plt.xlim(xx.min(), xx.max())

plt.title('SVC with Linear Kernel ')
```
