# Mean Shift Algorithm

This is a very popular and powerful clustering algorithm used in unsupervised learning. It does not make use of parameters so it is a non-parametric algorithm. 

It is also called hierarchical clustering or mean shift cluster analysis.

```python
#import the necessary packages
import numpy as np
from sklearn.cluster import MeanShift
%matplotlib inline 
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
```

## Dataset

Here we will generate a 2-dimensional dataset, containing 4 blobs by making use of the make_blob object from the sklearn.dataset package

```python
from sklearn.datasets._samples_generator import make_blobs
```

```python
#building and visualizing the dataset
centers = [[2,2], [4,5], [3,10]]
x,y= make_blobs(n_samples=500, centers=centers, cluster_std=1)
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
#instantiating the model
model = MeanShift()
```

```python
model.fit(x)
labels=model.labels_
cluster_centers = model.cluster_centers_
```

```python
#output cluster centers and estimated number of clusters
print(cluster_centers)
n_clusters = len(np.unique(labels))
print("Estimated number of Clusters:", n_clusters)
```

```python
#plotting prediction and fit according to clusters found
colors = 10*['r.','g.','b.','c.','k.','y.','m.']
for i in range(len(x)):
    plt.plot(x[i][0], x[i][1], colors[labels[i]], markersize=10)
plt.scatter(cluster_centers[:,0], cluster_centers[:,1], marker="x", color='k', s=150, linewidths=5, zorder=10)
plt.show()
```
