# K-Means Clustering

This is one of the well known clustering algorithms. It is an iterative clustering algorithm. 

```python
#import necessary packages
from sklearn.cluster import KMeans
%matplotlib inline 
import matplotlib.pyplot as plt
```

## Dataset

Here we will generate a 2-dimensional dataset, containing 4 blobs by making use of the make_blob object from the sklearn.dataset package

```python
from sklearn.datasets._samples_generator import make_blobs

x, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.40, random_state=0)
```

```python
plt.scatter(x[:, 0], x[:, 1], s = 50)
plt.show()
```

```python
#initializing k-means algorithm
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(x)
y_kmeans = kmeans.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_

#plotting prediction and fit according to clusters found
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.4)
plt.show()
```
